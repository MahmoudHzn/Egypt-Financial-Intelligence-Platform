```python
import yfinance as yf
import pandas as pd

# Download daily gold futures data for 10 years
gold = yf.Ticker("GC=F")
df = gold.history(period="10y", interval="1d")

# Save to CSV
df.to_csv("gold_daily_data.csv")
print(df.head())
```
